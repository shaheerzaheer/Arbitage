```python
import requests 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
x=requests.get('https://api.cryptonator.com/api/full/btc-usd').json() ['ticker']['markets']
x
```

```python
z=pd.DataFrame(x)
z
```

```python
market=np.array(z['market'].values)
price=np.array(z['price'].values)
volume=np.array(z['volume'].values)
q=np.max(price)
w=np.min(price)
e=float(q)-float(w)
print('max price',q)
print('min price',w)
print('max volume',np.max(volume))
print('min volume',np.min(volume))
print('Difference Between Prices is:' , e)
plt.scatter(price,volume)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=40
fig_size[1]=20
plt.rcParams.update({'font.size': 60})
for market,price,volume in zip(market,price,volume):
 plt.annotate(market,xy=(price,volume), xytext=(5,-5),textcoords='offset points', fontsize=40)
plt.title('Crypto Data', fontsize=60)
plt.xlabel('price',fontsize=60)
plt.ylabel('volume' , fontsize=60)
plt.xticks(rotation=70)
plt.show()
```
